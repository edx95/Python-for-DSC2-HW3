#### 1. Для чего и в каких случаях полезны различные варианты усреднения для метрик качества классификации: micro, macro, weighted?

Micro

    Micro усреднение - это метрика accuracy - предпочтительно использовать когда в данных нет дисбаланса классов. 

Macro

    Macro усреднение - расчитывается как среднеарифметическая метрик по всем классам, предпочтительно использовать когда есть дисбаланс классов и важны верные класификации в каждом классе.

Weighted

    Weighted  - взвешенное усреднение - расчитывается как взвешенное среднее, предпочтительно его использовать когда необходимо учитывать важность класса в зависимости от колличества объектов с таким классом.



#### 2. В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?

XGBoost - eXtreme Gradient Boosting, т.е. в основе данной модели лежит градиентный бустинг:

    подсчитывает похожеть между элементами и производит стрижку результирующих деревьев, основываясь на гиперпараметре гамма;
    прирост информации в XGBoost подсчитывается, как разница между суммой веток и вершины;
    для выбора разбиения использует сортировку и модели, основанные на анализе гистограмм (алгоритм разбиения деревьев);
    не умеет обрабатывать катекориальные переменные (обработка категориальных переменных по каждому алгоритму);
    средняя скорость работы алгоритма.

LightGBM - облегченная версия градиентного бустинга:

    заостряет внимание на ошибках и не использует всю выборку;
    группирует разреженные признаки, типичные для бинарного формата машинного обучения, а также признаки с близкими диапазонами значений;
    для выбора критерия разбиения используется техника односторонней выборки на основе градиента Gradient-based One-Side Sampling (GOSS, алгоритм разбиения деревьев)
    умеет работать с числовыми категориальными переменными, со строковыми нет (обработка категориальных переменных по каждому алгоритму);
    самая быстрая скорость работы алгоритма.

Catboost - быстрая модель градиентного бустинга, построенная на симметричных разветвлениях:

    выдерживая симметрию, модель решает проблемы классификации быстрее аналогов;
    модель сама переводит признаки в нужный формат без использования функции get_dummies и имеет встроенные алгоритмы, препятствующие переобучению
    умеет работать со строковыми и числовыми категорийными переменными (обработка категориальных переменных по каждому алгоритму);
    самая медленная скорость работы алгоритма.
